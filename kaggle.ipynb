```python
books = pd.read_csv("/kaggle/input/amazon-books-reviews/books_data.csv")
books
```

```python
books = books[['Title','description','authors','image','previewLink','categories','ratingsCount']]
books.dropna(inplace=True)
```

```python
books
```

```python
import ast

# Convert stringified lists into actual lists
books['categories'] = books['categories'].apply(ast.literal_eval)

# Flatten the list and find unique categories
unique_categories = set([item for sublist in books['categories'].tolist() for item in sublist])
```

```python
select_categories = [title for title in unique_categories if 'fiction' in title.lower()]
```

```python
select_categories
```

```python
# Create a boolean mask
mask = books['categories'].apply(lambda x: any(item for item in x if item in select_categories))

# Apply the mask to the dataframe
filtered_books = books[mask]
```

```python
filtered_books
```

```python
final_bookList = filtered_books['Title'].to_list()
```

```python
final_bookList
```

```python
selectedReviewsFinal = selectedReviews[selectedReviews['Title'].isin(final_bookList)]
```

```python
len(final_bookList)
```

```python
def evaluate_fraction(fraction):
    numerator, denominator = map(int, fraction.split('/'))
    if denominator != 0:
        return numerator / denominator
    else:
        return 0  # or some other value that makes sense in your context

selectedReviewsFinal['review/helpfulness'] = selectedReviewsFinal['review/helpfulness'].apply(evaluate_fraction)
```

```python
# selectedReviewsFinal
```

```python
# Sort the DataFrame by review/helpfulness score in descending order
selectedReviewsFinal = selectedReviewsFinal.sort_values(by='review/helpfulness', ascending=False)

# Group the DataFrame by Title and select the top 50 reviews for each group
# Create a new column to count occurrences of each title
selectedReviewsFinal['title_count'] = selectedReviewsFinal.groupby('Title').cumcount() + 1

# Filter rows where count is between 3 and 5
# next_three_reviews = selectedReviewsFinal[(selectedReviewsFinal['title_count'] >= 3) & (selectedReviewsFinal['title_count'] <= 7)]

# Drop the helper column as it's no longer needed
# next_three_reviews = next_three_reviews.drop(columns=['title_count'])
selectedReviewsFinal = selectedReviewsFinal.groupby('Title').head(2)

# Print the resulting DataFrame with the selected reviews
selectedReviewsFinal
```
